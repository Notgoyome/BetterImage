```python
# def apply_random_transformations(image):
#     transformations = {}

#     # Modifier la luminosité
#     enhancer = ImageEnhance.Brightness(image)
#     brightness_factor = random.uniform(0.5, 1.5)
#     image = enhancer.enhance(brightness_factor)
#     transformations['brightness'] = brightness_factor

#     # Modifier la saturation
#     enhancer = ImageEnhance.Color(image)
#     saturation_factor = random.uniform(0.5, 1.5)
#     image = enhancer.enhance(saturation_factor)
#     transformations['saturation'] = saturation_factor

#     # Modifier le contraste
#     enhancer = ImageEnhance.Contrast(image)
#     contrast_factor = random.uniform(0.5, 1.5)
#     image = enhancer.enhance(contrast_factor)
#     transformations['contrast'] = contrast_factor

#     # Modifier la netteté
#     enhancer = ImageEnhance.Sharpness(image)
#     sharpness_factor = random.uniform(0.5, 1.5)
#     image = enhancer.enhance(sharpness_factor)
#     transformations['sharpness'] = sharpness_factor

#     return image, transformations

def create_transformed_dataset(input_folder, output_folder, csv_output):
    os.makedirs(output_folder, exist_ok=True)
    images = os.listdir(input_folder)
    
    with open(csv_output, 'w', newline='') as csvfile:
        fieldnames = ['image_path', 'brightness', 'saturation', 'contrast', 'sharpness'] # Les colonnes du CSV
        #fieldnames = ['image_path', 'brightness', 'saturation', 'contrast', 'sharpness', 'temperature'] # Les colonnes du CSV
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        for image_name in images:
            if image_name.endswith('.jpg') or image_name.endswith('.png'):
                image_path = os.path.join(input_folder, image_name)
                # Appliquer les transformations
                transformed_image, transformations = apply_random_transformations(image_path)

                # Sauvegarder l'image transformée
                transformed_image_path = os.path.join(output_folder, f"transformed_{image_name}")
                transformed_image.save(transformed_image_path)

                # Écrire les transformations dans le CSV
                transformations['image_path'] = transformed_image_path
                writer.writerow(transformations)

                print(f"Image transformée sauvegardée: {transformed_image_path}")
```

```python
input_folder = 'raw_anime'
output_folder = 'dataset_anime'
csv_output = 'dataset_anime/transformations.csv'

create_transformed_dataset(input_folder, output_folder, csv_output)
```
